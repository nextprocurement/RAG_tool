```python
import pandas as pd
import pathlib
```

```python
path_objectives = pathlib.Path("/export/data_ml4ds/NextProcurement/PLACE/pliegos_objectives/all_extracted_12aug_es_acr")

file_count = sum(1 for file in pathlib.Path(path_objectives).rglob('*') if file.is_file())
all_dfs = []
for el, path in enumerate(pathlib.Path(path_objectives).rglob('*')):
    print("*" * 50)
    print(f"-- -- Processing file {el+1} / {file_count}: {path}")
    print("*" * 50)

    df = pd.read_excel(path)
    all_dfs.append(df)

df_all = pd.concat(all_dfs)
df_all
```
